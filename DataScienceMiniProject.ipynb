```python
DF=pd.read_csv('time_series_covid_19_confirmed.csv')
DF.set_index('Country/Region',inplace=True)
DF=DF.groupby('Country/Region').sum()
data2=data.copy()
data2.reset_index(inplace=True)
#print(data2)
data2=data2.groupby('index').sum()
print(data2)
print(DF.shape)
print(data2.shape)
#print(data2.index)
```

```python

#print(DF.index)
#print(data2.index)
for indx in data2.index :
    if indx in DF.index :
       # print('Found')
        continue
    else :
       # print('Not found')
        data2.drop(indx,inplace=True)
#print(DF.index)
#print(data2.index)
print(data2.shape)
print(DF.shape)
```

```python
data2['Lat']=""
data2['Long']=""
#print(data2.index)
#print(DF['Lat']['Afghanistan'])
for index in data2.index :
    if(index in data2.index):
       # print(index + ": " + str(DF['Lat'][index]))
        data2.at[index, 'Lat'] = DF['Lat'][index]
        data2.at[index, 'Long'] = DF['Long'][index]
data2.tail(10)
```

```python
world=folium.Map()
world
```

```python
for i in range(0,len(data2.index)):
    folium.Circle(
        location=[data2.iloc[i]['Lat'],data2.iloc[i]['Long']],
        tooltip = "<h5>"+data2.iloc[i].name+"</h5>"+
        "<li>Max_Infection_Rate:"+str(data2.iloc[i]['Max_Infection_Rate'])+"</li>"+
        "<li>GDP per capita:"+str(data2.iloc[i]['GDP per capita'])+"</li>"+
        "<li>Social support:"+str(data2.iloc[i]['Social support'])+"</li>",
        
        radius = (int((data2.iloc[i]['Max_Infection_Rate']))*2),
        color = '#ff6600',
        fill_color = '#ff6600',
        fill=True).add_to(world)
world
```

The above graph contains the data for Change in the 'GDP per capita' along y-axis with respect to the 'Max_Infection_Rate' along x-axis for various countries affected by covid_19. The data in the graph is based on the covid19 cases from january 2020 to 12th of september 2020. Due to covid_19 the GDP per capita for maximum countries were highly affected with only a small number of max_infection_rate as we can se from the above graph. There are very few countries whose GDP remain unaffected even with the highest maximum infection rate.
